```python
import bs4 as bs
import urllib.request
import re
import nltk
nltk.download('punkt')
import sys
```

```python
wiki_page = urllib.request.urlopen("https://en.wikipedia.org/wiki/India")
wiki = wiki_page.read()
```

```python
parse_wiki = bs.BeautifulSoup(wiki, "lxml")
article = parse_wiki.find_all("p")
```

```python
text = ""

for i in article:
  text += i.text
  
```

```python
text
```

```python
# clean the extracted text

text = re.sub(r"\[[0-9]*\]", " ", text)
text = re.sub(r"\s", " ", text)
new_text = re.sub("[^a-zA-Z]", " ", text)
new_text = re.sub(r"\s", " ", new_text)
```

```python
new_text
```

```python
# tokenize paragraphs into sentences

sentences = nltk.sent_tokenize(text)
nltk.download("stopwords")

```

```python
# tokenize sentences to words

stopwords = nltk.corpus.stopwords.words("english")

token_freq = {}
for token in nltk.word_tokenize(text):
  if token not in stopwords:
    if token not in token_freq.keys():
      token_freq[token] = 1
    else:
      token_freq[token] += 1
```

```python
# weighted frequency

max_freq = max(token_freq.values())

max_freq
```

```python
for token in token_freq.keys():
  token_freq[token] = (token_freq[token]/max_freq)
```

```python
weight = {}

for sent in sentences:
  for token in nltk.word_tokenize(sent.lower()):
    if token in token_freq.keys():
      if len(sent.split(" ")) < 30:
        if sent not in weight.keys():
          weight[sent] = token_freq[token]
        else:
          weight[sent] += token_freq[token]
```

```python
# HEAP QUEUE / PRIORITY QUEUE ALGORITHM

import heapq 

extracted_sentences = heapq.nlargest(5, weight, key = weight.get)
```

```python
summary = " ".join(extracted_sentences)
summary
```
